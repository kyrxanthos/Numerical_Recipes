```python
def fourier(signal):
    """
    Perfrom Fourier Transform.
    
    Parameters
    ----------
    signal : string
        NMR signal to use
        Options: [s1, s2, ... , s16]

    Returns
    -------
    y : array
        NRM signal
    freq : array
        Array of frequencies for amplitude spectrum
    scaled_hh : array
        Array of f_k (Amplitude spectrum)
    nn : float
        size of hh array
    """
    
    y = df[signal].values
    hh = np.fft.rfft(y)
    nn = hh.size
    freq = np.linspace(0, (nn-1)/tmax, nn)
    
    return y, freq, hh, nn
```

Filter out the noise by passing the signal through a bandpass filter centered at the water peak with a width $\pm$30 Hz. Plot the filtered signal y(t) for t=[0,0.5).

```python
def i_fourier(hh, nn, freq):
    """
    Perfrom Inverse Fourier Transform.
    
    Parameters
    ----------
    hh : array
        Array of f_k (Amplitude spectrum)
    nn : float
        size of hh array
    freq : array
        Array of frequencies for performed FT

    Returns
    -------
    y2 : array
        recovered NRM signal from IFT
    hh : array
        Filtered array of f_k's, it is 0 except near peak
    max_freq : float
        frequency at the peak

    """

    # index of the peak of the FFT
    imaxf = np.abs(hh).argmax()
    # remove everything within 30 (arbitrary frequency units)
    hh[:nn][np.abs(freq - freq[imaxf]) > 30] = 0
    max_freq = freq[imaxf]
    y2 = np.fft.irfft(hh)
    
    return y2, hh, max_freq
```

Determine $f_0$ in a different way: find the position of the water peak in the amplitude spectrum by fitting the curve

$

\tilde{y}_{\rm theor}(f) = C + \frac{A}{\sqrt{\lambda^4 + (f^2 - f_0^2)^2 + 2 \lambda^2(f^2 + f_0^2))}}
\end{align}
$

This curve comes from Fourier-transforming the exponentially damped sine function from the previous task, plus a constant C to account for background noise. $\lambda$ denotes the damping rate (inversely proportional to $t_0$ from task 4).

As before, use only the first of the 16 samples for this task. The frequency should be accurate to within 0.01 Hz of the correct answer.

```python
def fun(f, C, A, f0, lamda):
    y_theor = C + (A / (np.sqrt(lamda**4 + (f**2 - f0**2)**2 
                            + 2 * lamda**2 * (f**2 + f0**2))))
    return y_theor
```
